```python
import numpy as np
def split(trace, n):       
    tab = {}
    for i in range(trace+1)[1:]:
        tab[i] = []
    tab[1] = [[1]]
    memo(trace, n, tab)
    return tab[trace]


def memo(tr, n, tab):
        if(tab[tr] != []):
            return tab[tr]
        elif(tr == 1):
            return [1]
        else:
            for i in range(n+1)[1:-1]:
                ls = memo(tr-i, min(n,tr-i), tab)
                for case in ls:
                    if(len(case) <= n-1):
                        res = case + [i]
                        res.sort()
                        if(not(res in tab[tr])):
                            tab[tr].append(res)
                    return tab[tr]
    
split(22,5)
```
